```python
from google.colab import files
from ultralytics import YOLO
from IPython.display import Video, display

# Step 1: Upload your fire.mp4 from PC
uploaded = files.upload()   # only video, since best.pt is already in Colab

# Step 2: Load the model from Colab
model = YOLO("/content/best.pt")

# Step 3: Run detection on uploaded video
results = model.predict(source="/content/9487-220088145_small.mp4", save=True)

# Step 4: Get the path of saved output video
output_path = results[0].save_dir + "/fire.mp4"

# Step 5: Display detected video inside Colab
display(Video(output_path, embed=True))
```
